# Regression Models

## Happiness Score

### All Variables Happiness Score Regression

```python
features = final_df[['GDP_per_Capita','Generosity','GINI Coefficient','GDP','GDP_per_Capita','Family','Life Expectancy','Freedom']]
target = final_df['Happiness_Score']

features = sm.add_constant(features)

model = sm.OLS(target, features)
results = model.fit()

print(results.summary())

print("Regression Coefficients:")
for i, feature in enumerate(features.columns):
    if i == 0:
        print(f"Intercept: {results.params[i]}")
    else:
        print(f"{feature}: {results.params[i]}")

print("\nP-values:")
print(results.pvalues)
```

### Best Happiness Score Regression Model

```python
features = final_df[['GDP', 'Generosity', 'GINI Coefficient', 'GDP', 'GDP_per_Capita', 'Family', 'Life Expectancy', 'Freedom']]
target = final_df['Happiness_Score']

features = add_constant(features)

# Stepwise Regression
def stepwise_regression(X, y):
    remaining_features = list(X.columns)
    best_score = -float('inf')
    best_model = None

    for k in range(1, len(X.columns) + 1):
        for subset in itertools.combinations(remaining_features, k):
            model = sm.OLS(y, X[list(subset)]).fit()
            score = model.rsquared_adj

            if score > best_score:
                best_score = score
                best_model = model
                best_subset = subset

        remaining_features = list(set(remaining_features) - set(best_subset))

    return best_model

model = stepwise_regression(features, target)

print(model.summary())

print("Regression Coefficients:")
for i, feature in enumerate(model.params.index):
    if i == 0:
        print(f"Intercept: {model.params[i]}")
    else:
        print(f"{feature}: {model.params[i]}")

print("\nP-values:")
print('BASED ON THE STEPWISE REGRESSION')
print(model.pvalues)
```

### Best Dystopia Distance Regression Model

```python
final_dataframe = pd.concat([full_2020,full_2021], ignore_index = True, join = 'inner')
final_dataframe.rename(columns = {'year': 'Year'}, inplace = True)
final_dataframe.set_index('Country', inplace = True)
final_dataframe['GDP'] = final_dataframe['GDP'].str.replace('$', '', regex = True)
final_dataframe['GDP'] = final_dataframe['GDP'].str.replace(',', '', regex = True)
final_dataframe['GDP'] = final_dataframe['GDP'].astype(str).apply(lambda x: x.split('.')[0])
final_dataframe['GDP'] = np.int64(final_dataframe['GDP'])
final_dataframe.rename(columns = {'Happiness Score': 'Happiness_Score', 'GDP per Capita':'GDP_per_Capita'}, inplace = True)
final_dataframe.head()
```

```python
features = final_dataframe[['Generosity', 'Happiness_Score', 'GINI Coefficient', 'GDP', 'GDP_per_Capita', 'Family', 'Life Expectancy', 'Freedom', 'Perceptions of Corruption']]
target = final_dataframe['Dystopia Residual']

features = add_constant(features)

# Stepwise Regression
def stepwise_regression(X, y):
    remaining_features = list(X.columns)
    best_score = -float('inf')
    best_model = None

    for k in range(1, len(X.columns) + 1):
        for subset in itertools.combinations(remaining_features, k):
            model = sm.OLS(y, X[list(subset)]).fit()
            score = model.rsquared_adj

            if score > best_score:
                best_score = score
                best_model = model
                best_subset = subset

        remaining_features = list(set(remaining_features) - set(best_subset))

    return best_model

model = stepwise_regression(features, target)

print(model.summary())

print("Regression Coefficients:")
for i, feature in enumerate(model.params.index):
    if i == 0:
        print(f"Intercept: {model.params[i]}")
    else:
        print(f"{feature}: {model.params[i]}")

print("\nP-values:")
print('BASED ON THE STEPWISE REGRESSION')
print(model.pvalues)
```
